```python
import pandas as pd

from imblearn.over_sampling import SMOTE

pd.set_option('display.max_rows', 30)
pd.set_option('display.max_columns', 50)
pd.set_option('display.width', 1000)
```

```python
raw_data = pd.read_csv("raw/creditcard.csv")

# remove the class column as we want to learn unsupervised learning models
# raw_data = raw_data.drop(["Class"], axis=1)
```

```python
raw_data.shape
```

```python
raw_data.head()
```

```python
# oversample the dataset

x = raw_data.drop(["Class"], axis=1)
y = raw_data.filter(["Class"], axis=1)

smote = SMOTE()
x_over, y_over = smote.fit_resample(x, y)

oversampled_df = pd.DataFrame()
oversampled_df = pd.concat([x_over, y_over], axis=1)
# shuffle dataframe
oversampled_df = oversampled_df.sample(frac=1, random_state=42)
```
